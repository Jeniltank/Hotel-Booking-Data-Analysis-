#  Data Analysis  and Visualization


```python
import matplotlib.pyplot as plt

cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

legend_labels, _ = ax1.get_legend_handles_labels()

plt.title("Reservation status in different hotels")
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')

plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)

```

```python
City_hotel = df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotels', fontsize=30)

# Plot for Resort Hotel
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='blue')

# Plot for City Hotel
plt.plot(City_hotel.index, City_hotel['adr'], label='City Hotel', color='orange')

plt.xlabel('Reservation Status Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)  # Optional: adds a nice grid
plt.show()
```

```python


df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df)

plt.title('Reservation Status per Month', fontsize=20)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Number of Reservations', fontsize=15)
plt.legend(title='Is Canceled', labels=['Not Canceled', 'Canceled'])
plt.show()

```

```python
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate (ADR) per Month', fontsize=30)

sns.barplot(
    x='month',
    y='adr',
    data=df[df['is_canceled'] == 1].groupby('month')[['adr']].mean().reset_index(),
    palette='Blues'
)

plt.xlabel('Month', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(10,10))  # slightly bigger for better visibility
plt.title('Top 10 Countries with Reservation Cancellations', fontsize=18)

plt.pie(
    top_10_country,
    labels=top_10_country.index,
    autopct='%.2f%%',
    startangle=90,
    textprops={'fontsize': 12},  # make labels readable
    wedgeprops={'edgecolor': 'black'}  # clean separation
)

plt.tight_layout()  # ensures nothing gets cut off
plt.show()
```

```python
# Split the data into cancelled and not-cancelled bookings
cancelled_df = df[df['is_canceled'] == 1]
not_cancelled_df = df[df['is_canceled'] == 0]

# Group by date and calculate mean ADR for cancelled bookings
cancelled_df_adr = cancelled_df.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Group by date and calculate mean ADR for not-cancelled bookings
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the data
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate (ADR) Over Time', fontsize=16)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled', color='green')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled', color='red')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ensure the date column is in datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Filter data between 2016 and 2017
df_filtered = df[(df['reservation_status_date'] >= '2016-01-01') & 
                 (df['reservation_status_date'] <= '2017-12-31')]

# Split the data into cancelled and not-cancelled bookings
cancelled_df = df_filtered[df_filtered['is_canceled'] == 1]
not_cancelled_df = df_filtered[df_filtered['is_canceled'] == 0]

# Group by date and calculate mean ADR for cancelled bookings
cancelled_df_adr = cancelled_df.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Group by date and calculate mean ADR for not-cancelled bookings
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the data
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate (ADR) Over Time (2016–2017)', fontsize=16)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled', color='green')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled', color='red')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.grid(True)
plt.show()

```
